```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Create Celsius data
X = np.arange(-10, 41, 5) # From -10 to 40, in steps of 5

# 2. Create exact Fahrenheit conversion
y_true = X * 1.8 + 32

# 3. Add random noise to simulate real-world data
noise = np.random.normal(0, 5, size=y_true.shape) # Mean=0, StdDev=5
y_noisy = y_true + noise

# 4. Plot the data
plt.figure(figsize=(8, 6))
plt.scatter(X, y_noisy, color='blue', label='Noisy Data (Training Examples)')
plt.plot(X, y_true, color='red', linestyle='--', label='True Relationship')
plt.title('Supervised Learning: Celsius to Fahrenheit')
plt.xlabel('Celsius (Input Feature)')
plt.ylabel('Fahrenheit (Target Label)')
plt.legend()
plt.grid(True)
plt.show()

print("Input Celsius (X):\n", X)
print("\nTarget Fahrenheit (y):\n", y_noisy)
```
